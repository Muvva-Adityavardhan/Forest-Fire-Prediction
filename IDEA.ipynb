```python
import pandas as pd

# Load the dataset
file_path = 'C:/Users/muvva/OneDrive/Desktop/Rp/2021.csv'
data = pd.read_csv(file_path)

# Display the first few rows and the data info to understand the structure
data.head(), data.info()




import matplotlib.pyplot as plt
import seaborn as sns

# Convert the date column to datetime format and extract the month
data['date'] = pd.to_datetime(data['date'], format='%d-%m-%Y')
data['month'] = data['date'].dt.month

# Count the number of forest fires for each month
monthly_counts = data['month'].value_counts().sort_index()

# Plot the distribution of forest fire occurrences by month
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_counts.index, y=monthly_counts.values, palette='autumn')
plt.title('Monthly Distribution of Forest Fire Occurrences in India')
plt.xlabel('Month')
plt.ylabel('Number of Forest Fires')
plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                        'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()




import geopandas as gpd
import folium
from folium.plugins import HeatMap

# Define a base map centered on India
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Prepare data for the heatmap: Filtered by a specific month (e.g., March), add lat, lon, and VPD
march_data = data[data['month'] == 3]  # Example: March

# Extract relevant columns for the heatmap, using `vpd_avg` as the intensity
heat_data = march_data[['lat', 'lon', 'vpd_avg']].values.tolist()

# Add the heatmap to the map
HeatMap(heat_data, radius=10, blur=15, max_zoom=1).add_to(m)

# Save map to HTML for viewing
map_path = "C:/Users/muvva/OneDrive/Desktop/Rp/India_Forest_Fire_Heatmap_2021.html"
m.save(map_path)

map_path

```

```python
import matplotlib.animation as animation
import numpy as np

# Initialize figure
fig, ax = plt.subplots(figsize=(8, 8))
sns.set_style("white")

# Prepare base map of India using latitude and longitude bounds
map_extent = [68, 98, 6, 38]
ax.set_xlim(map_extent[0], map_extent[1])
ax.set_ylim(map_extent[2], map_extent[3])
ax.set_title("Monthly Forest Fire Hot Zones in India")

# Create scatter plot for dynamic updates
scatter = ax.scatter([], [], s=10, alpha=0.6, c=[], cmap="hot", marker='o')

# Function to update the plot for each month
def update(month):
    ax.clear()
    ax.set_xlim(map_extent[0], map_extent[1])
    ax.set_ylim(map_extent[2], map_extent[3])
    ax.set_title(f"Forest Fire Hot Zones in India - Month {month}")

    # Filter data for the current month
    month_data = data[data['month'] == month]
    lat = month_data['lat'].values
    lon = month_data['lon'].values
    intensity = month_data['vpd_avg'].values  # Using vpd_avg as color intensity

    # Update scatter plot with new data
    scatter = ax.scatter(lon, lat, s=15, alpha=0.6, c=intensity, cmap="hot", marker='o')

# Generate animation
ani = animation.FuncAnimation(fig, update, frames=np.arange(1, 13), interval=1000)

# Save animation as a GIF
gif_path = 'C:/Users/muvva/OneDrive/Desktop/Rp/India_Forest_Fire_Hotzones_Animation.gif'
ani.save(gif_path, writer='pillow', fps=1)

gif_path
```
